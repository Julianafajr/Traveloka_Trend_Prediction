DATA CLEANING

```python
import pandas as pd

# Importing the data as df and repeating the cleaning steps
df = pd.read_csv("data.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Step 1: Clean column names
df.columns = df.columns.str.strip().str.replace(';', '', regex=False).str.replace(' ', '_')

# Step 2: Apply value cleaning to the date-related columns
# Update column names after renaming
date_cols = ['Booking_Date', 'Pickup_Date', 'Issuance_Date']

# Fix Indonesian month names again just to be safe
month_map = {
    "Januari": "January", "Februari": "February", "Maret": "March", "April": "April",
    "Mei": "May", "Juni": "June", "Juli": "July", "Agustus": "August",
    "September": "September", "Oktober": "October", "November": "November", "Desember": "December"
}

# Replace month names and remove unwanted symbols
for col in date_cols:
    # Replace Indonesian month names
    for indo, eng in month_map.items():
        df[col] = df[col].astype(str).str.replace(indo, eng, regex=False)

    # Remove semicolons and extra spaces
    df[col] = df[col].str.replace(';', ' ', regex=False)
    df[col] = df[col].str.replace('  ', ' ', regex=False).str.strip()

# Show cleaned column names and examples from each date column
cleaned_date_examples = {col: df[col].unique()[:10] for col in date_cols}
df.columns.tolist(), cleaned_date_examples
```

```python
# Convert cleaned date columns to datetime, coercing invalid entries to NaT
for col in ['Pickup_Date', 'Issuance_Date']:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Check how many rows are still NaT (i.e., failed to convert)
missing_after_parsing = df[['Booking_Date', 'Pickup_Date', 'Issuance_Date']].isna().sum()
missing_after_parsing
```

```python
# Clean and extract hour and minute from time columns
time_columns = ['Booking_Time', 'Pickup_Time', 'Issuance_Time']

for col in time_columns:
    # 1. Remove 'GMT' and ';', strip spaces
    df[col] = (
        df[col].astype(str)
        .str.replace('GMT', '', regex=False)
        .str.replace(';', '', regex=False)
        .str.strip()
    )
    
    # 2. Split into components (HH, MM, SS) using '.' separator
    time_parts = df[col].str.split('.', expand=True)
    
    # 3. Assign Hour and Minute columns (as integers if possible)
    df[col + '_Hour'] = pd.to_numeric(time_parts[0], errors='coerce')
    df[col + '_Minute'] = pd.to_numeric(time_parts[1], errors='coerce')

# Show resulting hour and minute columns
df[['Booking_Time_Hour', 'Booking_Time_Minute',
    'Pickup_Time_Hour', 'Pickup_Time_Minute',
    'Issuance_Time_Hour', 'Issuance_Time_Minute']].head()
```

```python
# Drop the original raw time string columns after extracting hour and minute
df.drop(columns=['Booking_Time', 'Pickup_Time', 'Issuance_Time'], inplace=True)
```

```python
# Re-extract year, month, and day ONLY from valid datetime columns
# Skip 'Booking_Date' since it's not a valid datetime

valid_date_columns = ['Pickup_Date', 'Issuance_Date']

for col in valid_date_columns:
    df[f'{col}_Year'] = df[col].dt.year
    df[f'{col}_Month'] = df[col].dt.month
    df[f'{col}_Day'] = df[col].dt.day

# Drop only the valid datetime columns
df.drop(columns=valid_date_columns, inplace=True)

# Show newly created date parts
created_columns = [col for col in df.columns if 'Pickup_Date_' in col or 'Issuance_Date_' in col]
df[created_columns].head()
```

```python
# Count how many rows contain '-' in each column
dash_counts = {
    col: (df[col].astype(str) == '-').sum()
    for col in df.columns
    if (df[col].astype(str) == '-').any()
}

# Convert to a DataFrame for better readability
import pandas as pd
dash_counts_df = pd.DataFrame.from_dict(dash_counts, orient='index', columns=['Rows_with_dash'])
dash_counts_df.sort_values(by='Rows_with_dash', ascending=False)

```

```python
# Drop columns where all values are just '-'
columns_all_dash = [
    col for col in df.columns
    if (df[col].astype(str) == '-').sum() == len(df)
]

# Drop those columns from the DataFrame
df.drop(columns=columns_all_dash, inplace=True)

# Show remaining columns after drop
df.columns.tolist()
```

```python
df.isna().sum()
```

```python
# Step 1: Drop 'Total Publish Price' since it's completely NaN
df.drop(columns=['Total_Publish_Price'], inplace=True)
# Use forward-fill (ffill) to fill missing driver-related information with the previous non-null value
df['Driver_Assigned_Time'].fillna(method='ffill', inplace=True)
df['Driver_name'].fillna(method='ffill', inplace=True)
df['Driver_Phone_Number'].fillna(method='ffill', inplace=True)
# Re-check for any remaining missing values
remaining_nans = df.isna().sum()
remaining_nans[remaining_nans > 0]
```
